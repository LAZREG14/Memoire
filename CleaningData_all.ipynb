```python
withoutSlash_n =[]
for s in all_data:
    try:
        x = re.sub("\n", " ", s)
    except:
        pass
    withoutSlash_n.append(x)
```

```python
withoutSlash =[]
for s in withoutSlash_n:
    try:
        x = re.sub("\\'", "'", s)
    except:
        pass
    withoutSlash.append(x)
```

```python
import spacy
nlp_sm = spacy.load("fr_core_news_sm")
nlp_md = spacy.load("fr_core_news_md")
nlp_lg = spacy.load("fr_core_news_lg")
```

```python
f1 = ouvrir_json("all_data/clean_all_data.json")
f11 = ouvrir_json("all_data/all_data.json")
f2 = ouvrir_json("Data/voc_glaff.json")
```

```python
print(len(f1),len(f11))
```

```python
loc=[]
for texte in f1:
    doc = nlp_sm(texte)
    for entity in doc.ents:
        if entity.label_ =="LOC":
            loc.append(entity.text)
print(len(set(loc)))
print(set(loc))
```

```python
loc2=[]
for texte in f1:
    doc = nlp_md(texte)
    for entity in doc.ents:
        if entity.label_ =="LOC":
            loc2.append(entity.text)
print(len(set(loc2)))
print(set(loc2))
```

```python
loc3=[]
for texte in f1:
    doc = nlp_lg(texte)
    for entity in doc.ents:
        if entity.label_ =="LOC":
            loc3.append(entity.text)
print(len(set(loc3)))
print(set(loc3))
```

```python
test = [mot for mot in loc if mot.lower() not in f2]
print(len(test))
test2 = [mot for mot in loc2 if mot.lower() not in f2]
print(len(test2))
test3 = [mot for mot in loc3 if mot.lower() not in f2]
print(len(test3))
```

```python
x  =[]
x2 =[]
x3 =[]


for mot in test:
    if mot.lower() not in test:
        x.append(mot)
for mot in test2:
    if mot.lower() not in test2:
        x2.append(mot)
for mot in test3:
    if mot.lower() not in test3:
        x3.append(mot)
print(len(x))
print(len(x2))
print(len(x3))
ecrire_json("lieux_all_data_apr√©sNettoyage_sm.json",x)
ecrire_json("lieux_all_data_apr√©sNettoyage_md.json",x2)
ecrire_json("lieux_all_data_apr√©sNettoyage_lg.json",x3)
```
