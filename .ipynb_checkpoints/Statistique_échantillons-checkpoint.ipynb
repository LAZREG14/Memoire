```python
import json
import re
from textatistic import Textatistic

def ouvrir_json(chemin):
    f = open(chemin, encoding="utf-8")
    toto = json.load(f)
    f.close()
    return toto
def ecrire_json(chemin, contenu):
    w    = open(chemin, "w", encoding="utf-8")
    w.write(json.dumps(contenu, indent=2, ensure_ascii=False))
    w.close()
```

```python
echantillon1 = ouvrir_json("Data/echantillon1.json").values()
echantillon2 = ouvrir_json("Data/echantillon2.json").values()
```

```python
paroles_echantillon1 = ""
echantillon1_stats = []
lyricsNull=[]
lyricsfull=[]
datenull =[]
datefull =[]
featuring =[]
artistes =[]
for child in echantillon1:
    for info in child:
        try:
            if info["release_date"] != None:
                datefull.append(info["release_date"])
                len(datefull)
            else:
                datenull.append(info["release_date"])
        except:
            pass
                    
        try:
            if info["lyrics"] != None:
                paroles      = info["lyrics"]
                lyrics = re.sub("\n", ". ", paroles)
                paroles_echantillon1+=lyrics

                lyricsfull.append(paroles)
                album        = info.get("album")
                artiste      = album["artist"]["name"]
                artistes.append(artiste)
                feat_info  = info["featured_artists"]
                for v in feat_info:
                    feat = v["name"]
                    featuring.append(feat)
            else:
                lyricsNull.append(paroles)
        except:
            pass
nbChar = Textatistic(paroles_echantillon1).char_count
mots = paroles_echantillon1.split()
nbMots = len(mots)
#nbMot = Textatistic(paroles_echantillon1).word_count
nbPhrase = Textatistic(paroles_echantillon1).sent_count
nbSyll = Textatistic(paroles_echantillon1).sybl_count 
echantillon1_stats.append({"statistique_echantillon1":{"nombre de chanson": (len(lyricsNull))+(len(lyricsfull)),
                                                        "nombre d'artistes (featuring inclus)": len(set(artistes))+(len(set(featuring))),
                                                        "nombre de chanson avec la date de réalisation": len(datefull),
                                                        "nombre de chanson sans la date de réalisation": len(datenull),
                                                        "nombre de chanson sans les paroles": len(lyricsNull),
                                                        "nombre de chanson avec les paroles": len(lyricsfull),
                                                        "nombre de caractères": nbChar,
                                                        "moyenne de caractères par chanson": nbChar/len(lyricsfull),
                                                        "nombre de mots": nbMots,
                                                        "moyenne de mots par chanson": nbMots/len(lyricsfull),
                                                        "nombre de phrases": nbPhrase,
                                                        "moyenne de phrase par chanson": nbPhrase/len(lyricsfull),
                                                        "nombre de syllabes": nbSyll,
                                                        "moyenne de syllabes par chanson": nbSyll/len(lyricsfull)}})

ecrire_json("infos&stats/echantillon1_stats.json",echantillon1_stats)
print(datefull)
print(featuring)
```

```python
infos_echantillon1 = []

for child in echantillon1:
    for info in child:
        try:
            titre        = info.get("full_title")
            date         = info.get("release_date")
            album        = info.get("album")
            album_name   = info["album"]["name"]
            artiste      = album["artist"]["name"]
            if info["lyrics"] != None:
                paroles      = info["lyrics"]
                lyrics = re.sub("\n", ". ", paroles) #remplacer les \n par (. )
                nbChar = Textatistic(lyrics).char_count
                mots = lyrics.split()
                nbMots= len(mots)
                #nbMot       = Textatistic(lyrics).word_count
                nbPhrase     = Textatistic(lyrics).sent_count
                nbSyll       = Textatistic(lyrics).sybl_count
                #nbPolysyll  = Textatistic(lyrics).polysylbword_count

                infos_echantillon1.append({"titre": titre,
                                           "date": date,
                                           "album_name": album_name,
                                           "artiste": artiste,
                                           "lyrics": lyrics,
                                           "statistique":{"nombre de caractères": nbChar,
                                                          "nombre de mots": nbMots,
                                                          "nombre de phrases": nbPhrase,
                                                          "nombre de syllabes": nbSyll}})

                
        except:
            pass
ecrire_json("infos&stats/infos_echantillon1.json",infos_echantillon1)
```
