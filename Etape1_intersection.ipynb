```python
def lire_fichier(chemin):
  f = open(chemin, "r", encoding="utf-8")
  chaine = f.read()
  f.close()
  return chaine
def ecrire_fichier(chemin, contenu):
  w = open(chemin, "w", encoding="utf-8")
  w.write(contenu)
  w.close()
```

```python
import spacy
nlp = spacy.load("fr_core_news_sm")
```

```python
loc_corpus1=lire_fichier("locations_corpus1.json")
loc_corpus2=lire_fichier("locations_corpus2.json")
```

```python
loc1=loc_corpus1.split(",")
loc2=loc_corpus2.split(",")
```

```python
import re
list_loc2_difference = []
for loc in loc2:
    if loc not in loc1:
        list_loc2_difference.append(loc)

print("Liste des lieux retrouvés dans echantillon2 et pas dans echantillon1 : \n \n",set(list_loc2_difference)) #tous les "locations" qui appraissent dans l'echantillion2 et pas dans l'echantillion1
print("-"*100, "\n")


lieux2_mini = []

for x in set (list_loc2_difference):
    lowercase = re.findall(r'\b[a-z]*',x)

    if lowercase[0] != (''):
        lieux2_mini.append(lowercase[0])
print("Liste des lieux retrouvés en minuscule :\n \n",set(lieux2_mini))
    
```

```python
import collections
def mot_frequent(text):
    counts = collections.Counter(text.split())
    return counts.most_common(100)
```
